```python
import matplotlib.pyplot as plt
%matplotlib inline 
#using this i neednot use plt.show 
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

Scatter

```python
#plotting using matplotlib
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter in 2D')
plt.savefig('Test.png')
```

# Plot

```python
y = x*x
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r',linestyle='dashed',linewidth=2,markersize=12)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
#plt.title('Sine Wave')
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sin Wave Plot')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cos Wave Plot')
```

Bar Plot

```python
x=[3,4,5]
y = [14,5,12]
x2=[13,9,11]
y2=[4,12,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

# Histograms

```python
a=np.array([22,3,12,45,17,8,19,23,15,10,3,19,3,12,17,35,33,32,31,14,15,])
plt.hist(a,bins=15)
plt.title('Histogram')
plt.show()
```

# Box Plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

# Piechart

```python
#data to plot
labels = 'python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['blue','yellow','red','pink']
explode  = (0.1,0,0,0)
#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True)
plt.axis('equal')
plt.show()
```
